```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
rnd_arr = np.random.randint(0, 75, 100).reshape(2, 50)
```

```python
plt.scatter(rnd_arr[0], rnd_arr[1])
```

```python
plt.scatter(rnd_arr[0], rnd_arr[1])
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.grid() 
```

```python
fig, ax = plt.subplots(1, 2, figsize = (13, 5))
ax[0].hist(rnd_arr[0], color='g', label = 'First 50')
ax[1].hist(rnd_arr[1], color='b', label = 'last50')
ax[0].legend()
ax[1].legend()
```

```python
titanic_df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
fare_data = titanic_df[titanic_df.Fare < 275]
fig, ax = plt.subplots(1, figsize = (13, 5))
ax.hist(fare_data.Fare, bins = 10, color='g', label = 'Fares')
ax.legend()
```
